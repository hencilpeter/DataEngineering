```python
import pandas as pd
import numpy as np
```

```python
scores = {'Zone': ['North','South','South',
                   'East','East','West','West','West'], 
          'School': ['Rushmore','Bayside','Rydell',
                     'Shermer','Shermer','Ridgemont','Hogwarts','Hogwarts'],             
          'Name': ['Jonny','Joe','Jakob', 
                   'Jimmy','Erik','Lam','Yip','Chen'], 
          'Math': [78,76,56,67,89,100,55,76],
          'Science': [70,68,90,45,66,89,32,98]}
df = pd.DataFrame(scores, columns = 
                  ['Zone', 'School', 'Name', 
                   'Science', 'Math'])

df
```

```python
#group by zone and school 
for(zone, gp) in df.groupby(by=["Zone"]):
    print(f' {zone}')
    print("========")
    for(school, gp) in gp.groupby('School'):
        print('       ',school)
        print(gp)
    
```

```python
#group by zone and school 
for(zone, gp) in df.groupby(by=["Zone"]):
    print(f' {zone}')
    print("========")
    for(school, gp) in gp.groupby('School'):
        print(school)
        #remove index
        print(gp[['Name', 'Science', 'Math']].to_string(index=False))
```

```python
#group by after selecting few columns. df['Zone'] is required as its
#not in the selected list
for zone, gp in df[['Name', 'Science','Math']].groupby(df['Zone']):
    print (zone)
    print(gp)
```

```python
#group by zone and school
for zone, gp in df[['School','Name', 'Science','Math']].groupby([df['Zone'],df['School']]):
    print (zone)
    print('-----')
    print(gp)
   
```

```python
#print the average for all teh students in each school
for School, gp in df[['School','Name', 'Science','Math']].groupby(df['School']):
    print(School)
    print(gp.apply(['average', 'min', 'max']))
   
```

```python
df.groupby('School').mean()
```

```python
#which school belongs to which zone 
df.groupby(['Zone','School']).mean()
```

```python
#print only science
#approach 1
df.groupby(['Zone','School']).mean()[['Science']]
```

```python
#only science column mean
df.groupby(['Zone','School'])[['Science']].mean()
```

```python
df.groupby(['Zone','School']).agg('Science').mean()
```

```python
df.groupby(['Zone','School']).agg({'Science':np.mean})
```

```python
df.groupby(['Zone','School']).agg({'Science':[np.mean, np.mean, np.max]})
```

```python
df.groupby(['Zone','School']).agg({'Science':[np.mean, np.min, np.max], 'Math': [np.mean, np.min, np.max]})
```

```python
#introduce new column
r = df.groupby(['Zone','School']).mean()
r['Average'] = (r['Science'] + r['Math'])/2
r
```

```python
r = df.groupby('School').mean()
r['Average'] = (r['Science'] + r['Math']) / 2
r
```

```python
df = pd.read_csv('csv_data.csv')
df.head()
```

```python
#group all the model by year
for year, gp in df.groupby(['year']):
    print(year)
    for make, gp in gp.groupby('make'):
        print('   ', make)
        print(gp['model'].to_string(index=False))
```
