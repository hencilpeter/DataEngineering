```python
#sorting rows and columns
```

```python
import pandas as pd
import numpy as np
```

```python
dict2 = { 
  'w': 2.5,
  'x': pd.Series(3.5, index=list(range(1000)), dtype='float'),
  'y': np.random.randn(1000),
  'z': np.random.randn(1000),
}
df = pd.DataFrame(dict2)
```

```python
df
```

```python
#axis = 0 -> based on rows
#sort index -> either columns or indexes (rows)
df.sort_index(axis=0, ascending=False)
```

```python
df.sort_index(axis=1, ascending=False)
```

```python
#sort by values 
df.sort_values(by=['z','y'], ascending=False)
```

```python
#z in descending and y as ascending
df.sort_values(by=['z','y'], ascending=[False, True])
```
